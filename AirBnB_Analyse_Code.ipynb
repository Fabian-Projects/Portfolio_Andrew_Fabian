# Daten Bereinigung

```python
airbnb_df.drop_duplicates()
```

```python
# Überblick bekommen
describe_spalten = ['price', 'minimum_nights', 'number_of_reviews', 'last_review', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
describe_spalten = airbnb_df[describe_spalten].describe()
print(describe_spalten)
```

```python
# Nun will ich eine Nr_ID um jeden eintrag in meiner Liste einer bestimmten Nummer zuordnen zu können:
airbnb_df["Nr_ID"] = range(1, len(airbnb_df) + 1)

# Flights_ID an die erstel stelle bekommen
cols = airbnb_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
airbnb_df = airbnb_df[cols]

airbnb_df.head(20)
```

```python
airbnb_df.columns
```

```python
# Null einträge summieren

airbnb_df = airbnb_df.replace("", pd.NA)
airbnb_df = airbnb_df.replace(" ", pd.NA)

airbnb_df.isna().sum()
```

```python
# fehlenden name und host_name einträge begutachten
Null_name_host_name = conn.execute("SELECT * FROM AB_NYC_2019 WHERE name IS NULL OR host_name IS NULL").fetchdf()

# Zeige den neuen DataFrame an
Null_name_host_name
```

```python
#fehlenden review einträge begutachten
Null_last_review_reviews_per_month = conn.execute("SELECT * FROM AB_NYC_2019 WHERE last_review IS NULL OR reviews_per_month IS NULL").fetchdf()

# Zeige den neuen DataFrame an
Null_last_review_reviews_per_month.head()
```

```python
# Prüfen, ob es einträge gibt, welche reviews haben, aber fehlende werte bei last_review oder reviews_per_month
abfrage_review = conn.execute("SELECT * FROM AB_NYC_2019 WHERE number_of_reviews < 0 AND (last_review IS NULL OR reviews_per_month IS NULL)").fetchdf()

abfrage_review
```

# Erste Visualisierungen

```python
px.line(x = airbnb_df["Nr_ID"], y = airbnb_df["price"])
```

```python
# einzelnen neighbourhoods
neighbourhood_unique = airbnb_df["neighbourhood"].unique()
print("unique: ", "\n", neighbourhood_unique)
```

```python
# Summe der Airbnb's in den Nachbarschaften neighborhood:
neighbourhood_size = airbnb_df.groupby("neighbourhood").size().reset_index(name='size')
fig_neighbourhood_size = px.bar(neighbourhood_size, x="neighbourhood", y = "size", title="Nachbarschaft summe der Airbnb's")
fig_neighbourhood_size
```

```python
# Gruppierung nach Nachbarschaft und Berechnung des Durchschnittspreises
neighbourhood_prices_mean = airbnb_df.groupby("neighbourhood")["price"].mean().reset_index()
neighbourhood_prices_mean["price"] = neighbourhood_prices_mean["price"].round(2)
fig_neighbourhood_prices_mean = px.bar(neighbourhood_prices_mean, x="neighbourhood", y="price", title="Durchschnittspreis nach Nachbarschaft")
fig_neighbourhood_prices_mean
```

```python
# Size neighborhood:
neighbourhood_group_size = airbnb_df.groupby("neighbourhood_group").size().reset_index(name='size')
fig_neighbourhood_group_size = px.bar(neighbourhood_group_size, x="neighbourhood_group", y = "size", title="Nachbarschaft Gruppen summe der Airbnb's")
fig_neighbourhood_group_size
```

```python
# Gruppierung nach Nachbarschaft und Berechnung des Durchschnittspreises
neighbourhood_prices_mean = airbnb_df.groupby("neighbourhood_group")["price"].mean().reset_index()
neighbourhood_prices_mean["price"] = neighbourhood_prices_mean["price"].round(2)
fig_neighbourhood_prices_mean = px.bar(neighbourhood_prices_mean, x="neighbourhood_group", y="price", title="Durchschnittspreis nach Nachbarschaft Gruppen")
fig_neighbourhood_prices_mean
```
